```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sab
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

# Drop Unnamed Column

```python
df = df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# Data Transformation
change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-oct","5-10")
```

# Gender Ditribution

```python
plt.figure(figsize=(5,5))
ax = sab.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#From the above chart we have analysed that the numbers of female in the data is more then the numbers of male
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,5))
sab.heatmap(gb,annot=True)
plt.show()
```

```python
# from the above char we have concluded that the education of the parents have impact on their martialstatus
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,5))
sab.heatmap(gb1,annot=True)
plt.title("Relationhip between parents education and student score")
plt.show()
```

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ["groupA", "groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(mlist)

plt.pie(mlist,labels=l, autopct = "%1.2f%%")
plt.title("Distribution of erthenic group")
plt.show()
```
